
## Mass balance on reactive system. Extent of the reaction

##### Extent of the reaction is defined as: 

$\Large\xi = \frac{n_i - n_{i,0}}{\nu_i}$

##### which is equivalent to the formula:

$\Large n_i = n_{i,0} + \nu_i\xi$, or




##### Remarks:

* $\Large\xi>0\ \textrm{or}\ \xi<0\ \textrm{or}\ \xi=0$


* $\Large[\xi]=\textrm{mol}$


* $\Large\textrm{accu = in - out + gen - con} => \Large n_{i,out} = n_{i,in} + \nu_i\xi$ 

---

##### Example 1:

A + 2B => 3C

$\nu_A = -1, \nu_B = -2, \nu_C = 3$

$ n_A = n_{A,0} -1\xi$

$ n_B = n_{B,0} -2\xi$

$n_C = n_{C,0} +3\xi$

---

##### Example 2:

A + 2B => 3C

initially: $n_{A,0} = 5,\ \ n_{B,0} = 4,\ \ n_{C,0} = 3$,

after the reaction: $n_{A} = 3.5,\ \ n_{B} = 1,\ \ n_{C} = 7.5$

Calculate the $\xi$ (xi).

```python
# A
xi = (3.5 - 5)/-1
print("A:", xi, "mol")

# B
xi = (1 - 4)/-2
print("B:", xi, "mol")

# C
xi = (7.5 - 3)/3
print("C:", xi, "mol")

```

##### Example 3:

A + 2B <=> 3C

initially: $n_{A,0} = 5,\ \ n_{B,0} = 4,\ \ n_{C,0} = 7$,

after the reaction: $n_{A} = 7,\ \ n_{B} = 8,\ \ n_{C} = 1$

Calculate the $\xi$ (xi).

```python
# A
xi = (7 - 5)/-1
print("A:", xi, "mol")

# B
xi = (8 - 4)/-2
print("B:", xi, "mol")

# C
xi = (1 - 7)/3
print("C:", xi, "mol")
```

---

At the output: $n_i = n_{i,0} + \nu_i\xi$

Note $\nu_i$ is negative or positive

```python
nA0 = 2
nB0 = 5
nC0 = 0

nA = 0

# nA = nA0 - 1*xi
# nB = nB0 - 2*xi
# nC = nC0 + 3*xi

xi = nA0 - nA
print("xi =",xi)

nB = nB0 - 2*xi
nC = nC0 + 3*xi

print("nB =", nB)
print("nC =", nC)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline

figure(figsize=(10,7))


nA0 = 2
nB0 = 5
nC0 = 0

xi = np.linspace(-1,4)
nA = nA0 - 1*xi
nB = nB0 - 2*xi
nC = nC0 + 3*xi

plt.xlabel(r"$\xi$, mol")
plt.ylabel(r"$n_i$, mol")
plt.plot(xi, nA, label='A')
plt.plot(xi, nB, label='B')
plt.plot(xi, nC, label='C')

# ni range
plt.plot((-1,4),(0,0),'r:')

# marks: xi=0 => initial state
plt.plot(0, nA0, 'ro', label="Initial state")
plt.plot(0, nB0, 'ro')
plt.plot(0, nC0, 'ro')

# marks: xi=2 => final state
xi=2
nA = nA0 - 1*xi
nB = nB0 - 2*xi
nC = nC0 + 3*xi
plt.plot(2, nA, 'ko', label="Final state")
plt.plot(2, nB, 'ko')
plt.plot(2, nC, 'ko')



plt.title("Reaction extent as the reaction coordinate")
plt.legend()
plt.grid()
```
